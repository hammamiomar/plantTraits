```python
import os
import sys
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import torch
# Get the project root directory
ROOT_DIR = os.path.abspath(os.path.join(os.getcwd(), os.pardir))

# Add the project root directory to the Python path
sys.path.append(ROOT_DIR)
```

```python
from src.dataset import plantDataset
from src.dataset import transform
```

```python
dataset = plantDataset(csv_file='../data/train.csv', image_dir='../data/train_images', transform=transform)
dataloader = torch.utils.data.DataLoader(dataset, batch_size=1, shuffle=False)
```

```python
image, input_data, target_data,img_path = next(iter(dataloader))

# Display the image
plt.imshow(np.transpose(image.squeeze().numpy(), (1, 2, 0)))
plt.title('Image')
plt.show()

# Print the input data
print("Input Data:")
print(input_data)

# Print the target data
print("Target Data:")
print(target_data)

print("Image Path:")
print(img_path)
```
